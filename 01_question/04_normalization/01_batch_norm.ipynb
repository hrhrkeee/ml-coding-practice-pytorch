# 01. Batch Normalization

Batch Normalizationは、ミニバッチごとに正規化を行う手法です。

## Forward

$$
\hat{x}_i = \frac{x_i - \mu_B}{\sqrt{\sigma_B^2 + \epsilon}}
$$

$$
y_i = \gamma \hat{x}_i + \beta
$$

ここで:
- $\mu_B$: バッチの平均
- $\sigma_B^2$: バッチの分散
- $\gamma, \beta$: 学習可能なパラメータ
- $\epsilon$: 数値安定性のための小さな値
