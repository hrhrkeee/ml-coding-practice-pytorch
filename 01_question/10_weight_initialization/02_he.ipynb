# 02. He Initialization

He初期化は、ReLU系の活性化関数に適した重み初期化手法です。

## He Uniform

$$
W \sim U\left[-\sqrt{\frac{6}{n_{in}}}, \sqrt{\frac{6}{n_{in}}}\right]
$$

## He Normal

$$
W \sim N\left(0, \sqrt{\frac{2}{n_{in}}}\right)
$$

ここで:
- $n_{in}$: 入力ユニット数

## Xavier初期化との違い

- Xavierは $n_{in} + n_{out}$ を使用
- Heは $n_{in}$ のみを使用
- ReLUの勾配特性（半分が0）を考慮
