# 01. MSE (Mean Squared Error)

平均二乗誤差 (Mean Squared Error, MSE) は、予測値と実際の値の差の二乗の平均を計算します。
回帰問題でよく使われる損失関数です。

## Forward

$$
\text{MSE} = \frac{1}{N} \sum_{i=1}^{N} (y_{\text{pred},i} - y_{\text{true},i})^2
$$

## Backward

$$
\frac{\partial L}{\partial \text{pred}_i} = \frac{2}{N} (\text{pred}_i - \text{target}_i)
$$

ただし、`reduction='mean'`の場合です。
