```python
import numpy as np
from notebook_setup import test

@test("01_activation_function.test_04_softmax", filter_name="forward")
def softmax_forward(x: np.ndarray) -> np.ndarray:
    """
    Parameters
    ----------
    x : np.ndarray : shape (N, C)

    Returns
    -------
    result : np.ndarray : shape (N, C)
    """
    result = None

    # ここにコードを記述
    # ---------------------------------------- #
    # 数値安定化のために行ごとに最大値を引く
    shifted = x - np.max(x, axis=1, keepdims=True)
    exp_x = np.exp(shifted)
    result = exp_x / np.sum(exp_x, axis=1, keepdims=True)
    # ---------------------------------------- #

    return result
```

```python
import numpy as np
from notebook_setup import test

@test("01_activation_function.test_04_softmax", filter_name="backward")
def softmax_backward(x: np.ndarray, grad_output: np.ndarray) -> np.ndarray:
    """
    Parameters
    ----------
    x           : np.ndarray : shape (N, C)
    grad_output : np.ndarray : shape (N, C)

    Returns
    -------
    result : np.ndarray : shape (N, C)
    """
    result = None

    # ここにコードを記述
    # ---------------------------------------- #
    # 便利のためにまず forward を計算
    s = softmax_forward(x)
    # 各行ごとに (J_s * grad_output) を効率的に計算
    dot = np.sum(grad_output * s, axis=1, keepdims=True)
    result = s * (grad_output - dot)
    # ---------------------------------------- #

    return result
```
