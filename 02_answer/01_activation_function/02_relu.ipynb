```python
import numpy as np
from notebook_setup import test

@test("01_activation_function.test_02_relu", filter_name="forward")
def relu_forward(x: np.ndarray) -> np.ndarray:
    """
    Parameters
    ----------
    x : np.ndarray : shape (N, ...)

    Returns
    -------
    result : np.ndarray : shape (N, ...)
    """
    result = None

    # ここにコードを記述
    # ---------------------------------------- #
    result = np.maximum(0, x)
    # ---------------------------------------- #

    return result
```

```python
import numpy as np
from notebook_setup import test

@test("01_activation_function.test_02_relu", filter_name="backward")
def relu_backward(x: np.ndarray, grad_output: np.ndarray) -> np.ndarray:
    """
    Parameters
    ----------
    x           : np.ndarray : shape (N, ...)
    grad_output : np.ndarray : shape (N, ...)

    Returns
    -------
    result : np.ndarray : shape (N, ...)
    """
    result = None

    # ここにコードを記述
    # ---------------------------------------- #
    result = grad_output * (x > 0).astype(x.dtype)
    # ---------------------------------------- #

    return result
```
