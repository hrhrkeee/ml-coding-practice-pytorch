```python
import numpy as np
from notebook_setup import test

@test("01_activation_function.test_03_tanh", filter_name="forward")
def tanh_forward(x: np.ndarray) -> np.ndarray:
    """
    Parameters
    ----------
    x : np.ndarray : shape (N, ...)

    Returns
    -------
    result : np.ndarray : shape (N, ...)
    """
    result = None

    # ここにコードを記述
    # ---------------------------------------- #
    result = np.tanh(x)
    # ---------------------------------------- #

    return result
```

```python
import numpy as np
from notebook_setup import test

@test("01_activation_function.test_03_tanh", filter_name="backward")
def tanh_backward(x: np.ndarray, grad_output: np.ndarray) -> np.ndarray:
    """
    Parameters
    ----------
    x           : np.ndarray : shape (N, ...)
    grad_output : np.ndarray : shape (N, ...)

    Returns
    -------
    result : np.ndarray : shape (N, ...)
    """
    result = None

    # ここにコードを記述
    # ---------------------------------------- #
    tanh_x = np.tanh(x)
    result = grad_output * (1.0 - tanh_x * tanh_x)
    # ---------------------------------------- #

    return result
```
